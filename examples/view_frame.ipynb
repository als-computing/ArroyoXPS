```python
import numpy as np
import matplotlib.pyplot as plt

# Load the .npy file
data = np.load('/home/dmcreynolds/data/2.npy')

# Plot the 2D data as a heatmap
plt.imshow(data, cmap='viridis', aspect='auto')
plt.colorbar()  # Add a color bar for reference
plt.title("2D Data Heatmap")
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.show()
```
